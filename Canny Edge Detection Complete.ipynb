```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../desktop/CED_Pic.jpg')
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(image=img, threshold1=60, threshold2=90)
```

```python
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=img, threshold1=0, threshold2=255)
```

```python
plt.imshow(edges)
```

```python
med_val = np.median(img)
```

```python
lower = int(max(0, 0.3* med_val))
```

```python
upper = int(min(255,0.7 * med_val))
```

```python
edges = cv2.Canny(image=img, threshold1=lower , threshold2=upper)
```

```python
plt.imshow(edges)
```

```python
blurred_img = cv2.blur(img,ksize=(5,5))
```

```python
edges = cv2.Canny(image=blurred_img, threshold1=lower , threshold2=upper)
```

```python
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=blurred_img, threshold1=lower , threshold2=upper+50)
plt.imshow(edges)
```

Note: Threshold has an upper and lower limit. Threshold1 is the lower limit while threshold2 is the upper. They determine what pixels are the edge. Any pixels below the lower limit are not included, while every pixel above the upper limit is. The pixels in between may get picked in order to make a smooth edge.

"Canny" is a method that is less likely to be fooled by noise compared to other methods.
